```python
# book the maximum profit # brute force approach
mylist = [10,3,7,14,13]
min=mylist[0]
max=mylist[0]
for i in range(0,len(mylist)):
  if min>mylist[i]:
    min = mylist[i]
    for j in range(i,len(mylist)):
      if max<mylist[j]:
        max=mylist[j]
print(max-min)
```

```python
list1 = [10, 3, 7, 14, 16]

max_profit = 0
a = 0
b = 1

while b < len(list1):
    if list1[a] < list1[b]:
        profit = list1[b] - list1[a]
        if profit > max_profit:
            max_profit = profit
    else:
        a+=1

    b += 1
print(max_profit)
```
