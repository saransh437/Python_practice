```python
#  View vs Copy
# View means modification in orignal data and copy means chnages in duplicare data
#  will not reflect the orignal data
```

```python
a = np.array([1,2,3,4,5,5,6,7,8,9])
a[3:6] =0
a
```

```python
a = np.array([1,2,3,4,5,5,6,7,8,9])
b= a[3:6].copy()
b[:]=0
a
```
