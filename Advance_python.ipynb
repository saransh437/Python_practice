```python
# pre-built functions of Numpy
# (1). Zeros()-> it will create an array in which all the values will be 0
# in either one dimensional or multi-dimensional.
a= np.zeros(3)
a
b = np.zeros((3,4))
```

```python
# (2). ones()-> it will create an array in which all the values will be 1
# in either one dimensional or multi-dimensional.
a= np.ones(3)
print(a)
b = np.ones((3,4))    # two element cant go in one dimension
print(b)
```

```python
b = np.ones((5,6))
print(b.size)
print(b.shape)
print(b.ndim)
```
