```python
#  View vs Copy
# View means modification in orignal data and copy means chnages in duplicare data
#  will not reflect the orignal data
```

```python
a = np.array([1,2,3,4])
b = a.view()

b[0] = 100

print(a)   # original array
print(b)   # view array
```

```python
a = np.array([1,2,3,4])
b = a.copy()

b[0] = 100

print(a)
print(b)
```
