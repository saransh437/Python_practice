```python
def apply_funcs(x, funcs):
    for f in funcs:
        x = f(x)
    return x

funcs = [
    lambda x: x + 1,
    lambda x: x * 2,
    lambda x: x - 3,
    lambda x: x ** 2
]

result = apply_funcs(2, funcs)

print(result)
```

```python
# 7. Write a Python program to flatten a nested list (list within lists) into a single list.
# b. Flattened list: [1, 2, 3, 4, 5, 6, 7, 8]
li = [1, [2, 3], [4, [5, 6], 7], 8]
flt = []
for i in li:
  if type(i)==list:
       for j in i:
          if type(j)==list:
            for k in j:
              flt.append(k)
          else:
            flt.append(j)
  else :
    flt.append(i)
print(flt)
```

```python
# map is a high order function which takes another function as an argument
# map(function_name , iterable)
# it is lazy
# memory efficient
```

```python
def cube(n):
  return n*n*n

c = map(cube,[3,4,5])
print(next(c))
print(next(c))
print(next(c))
```

```python
# generator is a function in python which is usedd to generate the value n demand.
# yield keyword
# used to generate the value one at the time.
```

```python
def sara():
  n=0
  m=1

  while True:
    yield n
    n,m=m,m+n

gen = sara()
for i in range(7):
  print(next(gen))
```
