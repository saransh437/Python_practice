```python
# abstraction -> used to hide to the unwanted details from the user 
# to achive abstraction we have create the abstract method ->
# is a method which is not implemented and we have to make abstract class which have atleast 
# one abstract method
# can never make an object of abstract class 
```

```python
from abc import ABC,abstractmethod
class RBI(ABC):                       # abstract class 
    document='adhar'                 # class variable

    @abstractmethod
    def interest(self):                  # abstract method only implemented in child classes 
        pass

class HDFC(RBI):
    def interest(self):
        print('hdfc ka 10% loan')
h1 = HDFC()
h1.interest()
```

```python
# class ,instance and static method
class PythonA25:
    trainer = 'Tushar Sir'

    def __init__(self,st_name ,st_marks):
        self.st_name=st_name
        self.st_marks=st_marks
    def std_info(self):
        print(self.trainer , self.st_marks)
    
    @staticmethod            # and work same for every object and will work as normal function 
    def batch_info():
        print('start in january')    



s1 = PythonA25('saransh',20)  
s1.std_info()   
s1.batch_info()
PythonA25.batch_info()
```

```python
# class ,instance and static method

# no need of object in class and staticmethod
class PythonA25:
    trainer = 'Tushar Sir'

    def __init__(self,st_name ,st_marks):
        self.st_name=st_name
        self.st_marks=st_marks
    def std_info(self):
        print(self.trainer , self.st_marks)
    
    @staticmethod            
    def batch_info():
        print('start in january') 
    @classmethod
    def trainer_info(cls):
        print('trainer is :',cls.trainer)       



class PythonA26(PythonA25):
    trainer = 'vinay sir'
    def __init__(self,pst_name ,pst_marks):
        super().__init__(pst_name ,pst_marks)
    

s1 = PythonA26('sarash',150)
s1.trainer_info()  

s1 = PythonA25('sar',120)
s1.trainer_info()  
```
