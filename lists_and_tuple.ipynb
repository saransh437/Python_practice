```python
list1 = [10, 3, 7, 14, 16]
max_profit = 0
a = 0
b = 1
while b < len(list1):
    if list1[a] < list1[b]:
        profit = list1[b] - list1[a]
        if profit > max_profit:
            max_profit = profit
    else:
        a+=1
    b += 1
print(max_profit)
```

```python
# 3. Write a Python program to reverse a list without reverse or slicing operator
numbers = [1, 2, 3, 4, 5]
st=0
end=len(numbers)-1
while st<end:
  numbers[st],numbers[end]=numbers[end],numbers[st]
  st+=1
  end-=1
print(numbers)
```

```python
# 4. Write a Python program to sort a list in ascending order without sort functions.
numbers=[5,4,3,2,1]
st=0
end=1
while(st < len(numbers)-1):
  if numbers[st]>numbers[end]:
    numbers[st],numbers[end]=numbers[end],numbers[st]

  end+=1
  if end==len(numbers):
    st+=1
    end=st+1
print(numbers)
```

```python
# 5. Write a Python program to remove duplicates from a list while maintaining the order
# of elements.
numbers=[1, 2, 2, 3, 4, 4, 5]
st=0
end=1
while(st < len(numbers)-1):
  if numbers[st]==numbers[end]:
    numbers.pop(st)
  end+=1
  if end==len(numbers):
    st+=1
    end=st+1
print(numbers)
```
